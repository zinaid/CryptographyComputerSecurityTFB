# HASH FUNCTIONS

A hash function is a mathematical function that takes input data of any size and produces a fixed-size output, called a hash or digest.

Why use hashes in cryptography?

To verify data integrity — ensure data hasn’t changed
To store passwords securely — never store plain text passwords
To authenticate messages — detect tampering via HMAC
To enable digital signatures — sign the hash of a message

Key properties of cryptographic hash functions:

Deterministic – same input → same hash
One-way (preimage resistant) – cannot recover input from hash
Collision-resistant – hard to find two inputs producing the same hash
Avalanche effect – small input change → completely different hash
Fixed length – hash length does not depend on input size

Cryptographic hash functions have evolved over decades. They fall into different families:

MD Family
Examples: MD2, MD4, MD5
Output size: 128 bits
Status: Fast but broken (MD5 collisions are trivial)
Use today: Only for checksums, never for security

SHA Family (Secure Hash Algorithm)
SHA-0 / SHA-1: Obsolete (weak to collisions)
SHA-2: Most widely used
SHA-224, SHA-256, SHA-384, SHA-512
Output: 256–512 bits
Strong, collision-resistant, secure for modern applications
SHA-3: Latest NIST standard
Built differently from SHA-2 (Keccak algorithm)
Provides 224–512 bit outputs, highly secure


BLAKE Family
BLAKE2 / BLAKE3
Extremely fast, highly secure, can produce variable-length output
Gaining popularity in modern cryptography systems

Rule of thumb: For most applications, use SHA-256 / SHA-512 or BLAKE2.

Why HASH functions

Data Integrity:
Verify that messages or files haven’t been modified. Example: download a file → check SHA-256 hash.
Authentication:
Using HMAC (hash + secret key) to ensure messages are from the correct sender.
Password Storage:
Hash passwords before storing. Even if a database leaks, original passwords remain hidden.
Digital Signatures:
Sign the hash of a message instead of the full message for efficiency.

```python
# SHA 256 example

import hashlib

message = b"Hello, Cryptography!"

hash_object = hashlib.sha256(message)
digest = hash_object.hexdigest()
print("SHA-256 Digest:", digest)
```

```python
# HMAC EXAMPLE (AUTHENTICATION)

import hmac

key = b"mysecretkey"
message = b"Important message"

hmac_digest = hmac.new(key, message, hashlib.sha256).hexdigest()
print("HMAC Digest:", hmac_digest)
```

Password Hashing (Secure Method)

Important: General hash functions (SHA-256, SHA-3) are too fast for password hashing.
Passwords need slow, adaptive hashing to prevent brute-force attacks.

### BCRYPT

```python
import bcrypt

# Hashing the password
def hash_password(password: str) -> bytes:
    salt = bcrypt.gensalt()  # automatically generates a secure salt
    hashed = bcrypt.hashpw(password.encode('utf-8'), salt)
    return hashed

# Verifying the password
def check_password(stored_hash: bytes, password: str) -> bool:
    return bcrypt.checkpw(password.encode('utf-8'), stored_hash)

# Example
password = "MySecurePass123"
stored_hash = hash_password(password)
print("Stored hash:", stored_hash)

login_password = "MySecurePass123"
print("Password verification:", check_password(stored_hash, login_password))
```

bcrypt automatically embeds the salt into the hash
Adaptive cost factor slows down brute-force attacks
Always use bcrypt or Argon2 for storing passwords
